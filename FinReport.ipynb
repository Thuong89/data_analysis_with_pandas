```python
#importing all the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the csv file
data=pd.read_csv('PopFinland.csv',sep=';',header=[0],skiprows=[0],encoding='latin-1')
```

# Dataset
  

```python
data
```

```python
#Removing all the columns which have % sign to mininize the number of columns for data analysis
data=data[data.columns[~data.columns.str.contains(r'\%$')]]
data
```

```python
data.columns
```

```python
data=data.T
data
```

```python
#resetting the index
data.reset_index(inplace=True)
```

```python
data.columns=data.loc[0]
```

```python
data=data.drop(0)
```

```python
data
```

```python
#Removing the extra space infront of the columns name
Region=data['Year'].str.replace(r'(\s[CPF][a-z].+)','').str.strip()
```

```python
Region
```

```python
#Removing the extra space infront of the columns name
Attribute=data['Year'].str.replace(r'.+(?=(\s[CPF][a-z].+))','').str.strip()
```

```python
Attribute
```

```python
data.isna().sum().sum()
```

```python
data['Attribute']=Attribute
```

```python
data['Region']=Region
```

```python
data.drop('Year',axis=1,inplace=True)
```

```python
data
```

```python
data2=data.reset_index()
data2
```

```python
#grouping the dataset by Citizens of foreign countries attribute. Here, we can see the number of foreigners coming to Finland's different regions in different years.
data2[data2.Attribute=="Citizens of foreign countries"].groupby("Region").sum()
```

```python

data2.sort_values(["Region","Attribute"]).groupby("Region").first()
```

```python
data2.head()
```

```python
# In this line, we are trying to check which attributes have maximum value during the year 2010. For example, Uusimaa region have persons with foreign background as maximum values while whole country have persons born abroad as maximum values which is shown below. This can be useful to check if more foreigners moved to Finland or more people were born abroad and so on.
data2.loc[data2[2010]==data2.groupby("Region")[2010].transform("max")][["Region","Attribute",2010]]
```

```python
# Grouped using attribute to check every locations specific data.
grouped=data2.groupby(['Attribute','Region']).sum()
```

```python
grouped.head()
```

```python
data2.Region.unique()
```

```python
grouped.loc["Citizens of foreign countries","MK01 Uusimaa"]
```

```python
# Plotting the graph to see the trend of foreigners coming to Uusimaa region.
grouped.loc["Citizens of foreign countries","MK01 Uusimaa"].plot()
```

```python
#importing all the necessary libraries for visualization
import ipywidgets as widgets
from ipywidgets import interact
import plotly.graph_objects as go
import cufflinks as cf
from plotly.offline import iplot,init_notebook_mode

init_notebook_mode(connected=True)
cf.go_offline(connected=True)
```

```python
# Defining a function 
def plot_region(att,reg):
    grouped.loc[att,reg].iplot()
```

```python
# Plotting the graph
interact(plot_region,att=data2.Attribute.unique(),reg=data2.Region.unique())
```

<b> Here, the graph is present with attribute and region box where the user can select the desired location and attribute. X-axis represents the years while y-axis is numerical value. When selected the whole country as region and citizens of foreign countries we can see in 2019 the number was 267,629 while in 2018 the number was 257,572. In this way, comparison can be made during different years for different regions. It's more convienent to look at the graph and see the trend than looking at the dataset which makes it confusing to figure out the trend.
