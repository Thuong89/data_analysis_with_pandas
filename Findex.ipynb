# Catch and load original dataset

```python
url='https://globalfindex.worldbank.org/sites/globalfindex/files/2018-08/Global%20Findex%20Database.xlsx'
```

```python
data=pd.read_excel(url,'Data')
```

```python
data.head()
```

```python
data.shape
#huge amount of columns!!!
```

```python
data.rename({'Unnamed: 0':'Year','Unnamed: 2':'Country'},axis='columns', inplace=True)
```

```python
data.head()
```

# Extract necessary data:

Mobile money account accross countries in year 2011, 2014, 2017

```python
# extract to take variables to analyse on Mobile users in handling with finance.
# want to put Year and country to index but since they are not unique variables
mo_1=data[data.columns[data.columns.str.contains(r'^Mobile')]]
mo_2=data[['Year','Country']]
mo=mo_2.join(mo_1)
mo
```

```python
mo.info()
```

## Clean data

```python
mo_1.dropna(how='any',inplace=True)
# now only 174 columns 
mo_1
```

Convert the unit value for easy readable percentage

```python
mo_1=mo_1*100
```

```python
mo_cl=mo_2.join(mo_1,how='inner')  # join by matching index, only chose match
mo_cl
```

<mark>now we have clean data without missing value 'mo_cl'

Realize that the columns names are cubersome:
- we can remove redundance 'Mobile money account', because all this part is about Mobile already
- we also see redundance and later become confusing with '()', so we decide to drop the text part in () of each columns name.
- to keep understanding, we notice before deleting:
  - young aldults are age 15-24, old aldult are 25+, the rest columns are just 15+
  - figure are ratio of confirmed using mobile. (1 means 100%)

```python
mo_cl.columns=mo_cl.columns.str.replace(r'(Mobile\smoney\saccount\,?\s)','')
```

```python
mo_cl.columns=mo_cl.columns.str.replace(r'\(.+\)','')
```

```python
mo_cl.head()
```

```python
# one column with empty space, which actually belong to total:
mo_cl.rename({' ':'Total'}, axis='columns',inplace=True)
```

```python
mo_cl.head()
```

## Tidying Data and analysis

We can see that both columns labels and Year, and Country are not meaningful as value to calculate, but they are meaningful as labels.


```python
mo_cl1=pd.melt(mo_cl,id_vars=['Year','Country'],var_name='group')
```

```python
mo_cl1.head()
```

### How many percentage people used Mobile for financial matters globally, and by each groups of user?

This part analyses basing on tidy data: mo_cl1 and then add extra tidy steps along the way if neccessary.

```python
d1=mo_cl1.groupby(['group'])['value'].describe().sort_values(
    ['mean','max'],ascending=False)
d1
```

### Globally total

```python
total=d1.loc['Total']
total
```

```python
# scatter plot
mo_cl.plot.scatter(x='Year',y='Total')
```

INTEFRETING:
 - max >72% people use Mobile for financial purpose, however there is some country in the world that report of no Mobile usage. 
 - The mean value 9.96% theorectically represents for the whole world situation which is supprising low. It suggests the big room for development of Mobile segment here.
 - This mean value is far away from max value probably because the amount of people in non-mobile usage country account for much bigger portion of the world populartion. 
 - The previous argument is supported when we see the standard deviation is quite high 17.2.
 - Scatter plot also let us see that our clean data only contains year 2014 and 2017. For some reason, year 2011 contains too much missing value. Data is distributed scatterly, skewing more to the minimum value.

### Analysis by user group

```python
sub=d1[d1.index!='Total']
sub
```

```python
sub.plot(kind='line',y=['mean','max','min'], title='Global User Group', rot=90,
        figsize=(10,6))
```

INTEPRETING:
 - The group secondary education or less has highest percentage of mobile users basing on mean value 14.40%, and max value 83.89%
 - The group out of labor force has lowest percentage of using mobile for finance.
 - Comparision can be made between pairs of income, gender, etc, example:
     - There is small different between male and female in using mobile in finance.
 - The same pattern can be observed for all groups. 
     - Blue line (for mean value) is closer to green line(min values).
     - at 75 percentiles, value is still far away from max value. Majority of data points fall far away from max value.
