```python
#importing all the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the csv file
data=pd.read_csv('PopFinland.csv',sep=';',header=[0],skiprows=[0],encoding='latin-1')
```

 # Dataset (Statistics Finland)
 Statistics Finland combines collected data with its own expertise to produce statistics and information services.
Founded in 1865, Statistics Finland is the only Finnish public authority specifically established for statistics. It produces the vast majority of Finnish official statistics and is a significant international actor in the field of statistics.

We collected the Immigrants and integration dataset from there vast database collections for data analysis purposes. This dataset collects information on foreigners visiting Finland throughout the years depending on the regions of the country as well as overall whole country. For data analysis purposes, we have gathered information on Citizens of foreign countries, Persons born abroad, Foreign-language speakers and Persons with foreign background depending on the regions.

<b>Here, we have removed the unnecessary columns using % sign as a pattern to remove all the columns which have % sign since these columns are not necessary for our data analysis purposes.</b>

```python
data=data[data.columns[~data.columns.str.contains(r'\%$')]]
data
```

```python
data.columns
```

```python
data=data.T
data
```

<b> Resetting the index </b>

```python
#resetting the index
data.reset_index(inplace=True)
data
```

```python
data.columns=data.loc[0]
data
```

<b> We have dropped the extra year row from the dataset below using drop().</b>

```python
data=data.drop(0)
data
```

```python
#Removing the extra space infront of the columns name
Region=data['Year'].str.replace(r'(\s[CPF][a-z].+)','').str.strip()
```

Here, we have replaced the words after the regions name using regex pattern and assigned it to 'Region'. Also, strip() is used to get rid of the extra space in the begining of every columns.

```python
Region
```

 Similar process is carried out below to get rid of regions using regex pattern and assigned it to 'Attribute'.

```python
#Removing the extra space infront of the columns name
Attribute=data['Year'].str.replace(r'.+(?=(\s[CPF][a-z].+))','').str.strip()
```

```python
Attribute
```

```python
data['Attribute']=Attribute
```

```python
data['Region']=Region
```

```python
data.drop('Year',axis=1,inplace=True)
```

```python
data
```

```python
data2=data.reset_index()
data2
```

```python
#grouping the dataset by Citizens of foreign countries attribute. Here, we can see the number of foreigners coming to Finland's different regions in different years.
data2[data2.Attribute=="Citizens of foreign countries"].groupby("Region").sum()
```

 <b> In above line of code, we are trying to find the citizens of foreign countries using groupby to group the data according to the regions. In the dataset, we can see that the number of citizens of foreign countries moving to Finland has been increasing yearly.</b>
 
 <b> Same data is collected below using sort_values.

```python

data2.sort_values(["Region","Attribute"]).groupby("Region").first()
```

```python
data2.head()
data2=data2.drop(columns='index')
data2
```

```python
# In this line, we are trying to check which attributes have maximum value during the year 2010. For example, Uusimaa region have persons with foreign background as maximum values while whole country have persons born abroad as maximum values which is shown below. This can be useful to check if more foreigners moved to Finland or more people were born abroad and so on.
data2.loc[data2[2010]==data2.groupby("Region")[2010].transform("max")][["Region","Attribute",2010]]
```

<i> In this line, we are trying to check which attributes have maximum value during the year 2010. For example, Uusimaa region have persons with foreign background as maximum values while whole country have persons born abroad as maximum values which is shown above. This can be useful to check if more foreigners moved to Finland or more people were born abroad and so on.

```python
data2.loc[data2[2019]==data2.groupby("Region")[2019].transform("max")][["Region","Attribute",2019]]
```

 <i> In this part, we tried to check the data for the year 2019. If we compare this data with the data from 2010, we can see that in 2010 whole country had persons born abroad as maximum value while in 2019 whole country had person with foreign background as maximum value. This shows that there has been significance increase in foreigners moving to Finland.

```python
# Grouped using attribute to check every locations specific data.
grouped=data2.groupby(['Attribute','Region']).sum()
grouped
```

```python
#importing all the necessary libraries for visualization
import ipywidgets as widgets
from ipywidgets import interact
import plotly.graph_objects as go
import cufflinks as cf
from plotly.offline import iplot,init_notebook_mode

init_notebook_mode(connected=True)
cf.go_offline(connected=True)
```

```python
# Defining a function 
def plot_region(att,reg):
    grouped.loc[att,reg].iplot()
```

```python
# Plotting the graph
interact(plot_region,att=data2.Attribute.unique(),reg=data2.Region.unique())
```

<b> Here, the graph is presented with attribute and region box where the user can select the desired location and attribute. X-axis represents the years while y-axis is numerical value. When selected the whole country as region and citizens of foreign countries we can see in 2019 the number was 267,629 while in 2018 the number was 257,572. In this way, comparisons can be made during different years for different regions. It's more convienent to look at the graph and see the trend than looking at the dataset which makes it confusing to figure out the trend.
